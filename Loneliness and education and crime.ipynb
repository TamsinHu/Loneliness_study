```python
import pandas as pd
import plotly.express as px
```

# Loneliness index vs education and crime

EDUCATION: measures the lack of attainment of skills  
CRIME: measures the risk of personal and material victimisation

```python
loneliness_imd = pd.read_csv('england/loneliness_imd.csv')
loneliness_imd.drop(6270, inplace=True)
imd = pd.read_csv('england/imd2015eng.csv')
full_loneliness = pd.read_csv('england/final_data.csv')
```

```python
loneliness_imd.columns
```

```python
desired_columns = [
    'lsoa11', 'year', 'lsoa_code', 'lsoa_name', 'district_code', 'district_name',
    'loneills',
    'imd_score', 'imd_rank', 'imd_decile',
    'education_score', 'education_rank',
       'education_decile',
    'crime_score', 'crime_rank', 'crime_decile', 'adult_skills_sub_score', 'adult_skills_sub_rank',
       'adult_skills_sub_decile', 'young_sub_score', 'young_sub_rank',
       'young_sub_decile'
]

loneliness_crime_edu = loneliness_imd.filter(desired_columns, axis=1)
print(loneliness_crime_edu.shape)
```

### Correlation

```python
loneliness_crime_edu.corr()
```

Negative correlations for all education values and for crime

## Deciles analysis


```python
fig2 = px.scatter(loneliness_crime_edu, y='loneills', x='imd_decile', title='IMD Score vs lonliness index')
fig2.show()
```

```python
fig = px.violin(loneliness_crime_edu, y="loneills", x='education_decile', box=True, # draw box plot inside the violin
                 # can be 'outliers', or False
               )
fig.show()
```

```python
fig = px.violin(loneliness_crime_edu, y="loneills", x='adult_skills_sub_decile', box=True, # draw box plot inside the violin
                 # can be 'outliers', or False
               )
fig.show()
```

```python
fig = px.violin(loneliness_crime_edu, y="loneills", x='crime_decile', box=True, # draw box plot inside the violin
                 # can be 'outliers', or False
               )
fig.show()
```

```python
lonliness_imd_no_outl = loneliness_crime_edu[loneliness_crime_edu.loneills < 30]
```

```python
fig = px.violin(lonliness_imd_no_outl, y="loneills", x='education_decile', box=True, # draw box plot inside the violin
                 # can be 'outliers', or False
               )
fig.show()
```
