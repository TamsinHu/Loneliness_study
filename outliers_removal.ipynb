### Remove outliers
Dataframe without outliers is saved as data_without_outliers.csv

```python
Q1 = data.quantile(q=0.25)
Q3 = data.quantile(q=0.75)
IQR = Q3-Q1
df = data[~((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]
df.to_csv('england/data_without_outliers.csv')
```

```python
df.plot.scatter(x='adult_skills_sub_rank', y='loneills')
```

```python

df.corr()
```

Outliers are correctly removed. Correlation remains similar. 1327 data points have been removed. Let's store them in a separate dataframe...

```python
dropped_data = data[~data.isin(df)].dropna()
```
